**Structurer les données: Regrouper les acteurs par film et créer d'une colonne par catégorie de profession**

```python
pivot_df = df.pivot_table(
    index=['tconst', 'original_title'],
    columns='category',                   # Transformer les valeurs uniques de 'category' en nouvelles colonnes
    values='primaryName',                 # Afficher les données de 'primaryName' dans les nouvelles colonnes
    aggfunc=lambda x: list(x)             # Agrèger les valeurs de 'primaryName' en les regroupant dans une liste
).reset_index()                           # Réinitialiser l'index

pivot_df
```

```python
#Supprimer les lignes dupliquées dans le DataFrame df en se basant sur la colonne tconst

unique_cols = df.drop_duplicates('tconst')[['tconst', 'titleType', 'original_title',
                                            'startYear', 'runtimeMinutes', 'genres',
                                            'averageRating', 'overview', 'poster_path']]

```

```python
# Fusionner les deux DataFrames
final_df = pd.merge(unique_cols, pivot_df, on='tconst', how='left')

final_df.head()
```

```python
#fusionner les colonnes 'actor' et 'actress' pour avoir une colonne 'acteurs/trices'
final_df['acteurs/trice'] = final_df['actor'] + final_df['actress']
```

```python
#selection des colonnes finales de la dataframe

final_df = final_df[['tconst', 'original_title_x', 'startYear',
       'runtimeMinutes', 'genres', 'averageRating', 'overview', 'poster_path', 'acteurs/trice',
       'director']]
final_df
```

```python
#renommer la colonne 'original_tittle_x'

final_df.rename(columns= {'original_title_x' : 'original_title'}, inplace= True)
```

**Traduire les descriptions en français**

```python
#installation de la bibliothèque transformers de Hugging Face

from transformers import MarianMTModel, MarianTokenizer
!pip install transformers
```

```python
# chargement du modèle pré-entraîné (une seule fois, reste en mémoire)
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokennizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# fonction pour traduire un texte
def translate_text(text):
  inputs = tokennizer(text, return_tensors="pt", padding=True, truncation=True)
  translated = model.generate(**inputs)
  return tokennizer.decode(translated[0], skip_special_tokens=True)

final_df["overview_fr"] = final_df["overview"].apply(translate_text)
```
